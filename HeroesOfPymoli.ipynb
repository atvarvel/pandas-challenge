```python
# Create new dataframe containing only screen names and ages
age_demo_df = reduced_purchase_data[['SN','Age']]
```

```python
# Find maximum age to use as the last bin
max_age = max(age_demo_df['Age'])
# Create bins for age groups
bins = [0, 9, 14, 19, 24, 29, 34, 39, max_age]
# Create labels for age groups
group_labels = ["<10","10-14","15-19","20-24","25-29","30-34","35-39","40+"]

# Add new column to age demographics dataframe displaying which age group the player belongs in
age_demo_df['Age Group'] = pd.cut(age_demo_df['Age'], bins, labels=group_labels)
```

## Top Spenders

* Run basic calculations to obtain the results in the table below


* Create a summary data frame to hold the results


* Sort the total purchase value column in descending order


* Optional: give the displayed data cleaner formatting


* Display a preview of the summary data frame



```python
# Create dataframe that is grouped by screen name and also aggregates the Price column by finding the sum for each player
spenders_df = purchase_data.groupby(['SN']).agg({'Price': sum})
# Reset the index
spenders_df = spenders_df.reset_index()
# Create new dataframe sorting the previous dataframe by Price, put the values in descending order, and only show the top 5
top_spenders_df = spenders_df.sort_values(by=['Price'], ascending=False).head(5)
# Reset the index
top_spenders_df = top_spenders_df.reset_index()
# Delete the old index column
del top_spenders_df['index']
# Rename the price column as Total Purchase Value
top_spenders_df = top_spenders_df.rename(columns={'Price':'Total Purchase Value'})
```

```python
# To find each value for the top 5 players, filter the original dataframe by finding only the rows that contain that players name.
# Use that filtered dataframe to calculate each player's purchase count, average, and total.
spender_one_df = purchase_data[purchase_data['SN'] == top_spenders_df.iloc[0]['SN']]
spender_one_count = len(spender_one_df)
spender_one_avg = spender_one_df['Price'].mean()
spender_one_total = spender_one_df['Price'].sum()

spender_two_df = purchase_data[purchase_data['SN'] == top_spenders_df.iloc[1]['SN']]
spender_two_count = len(spender_two_df)
spender_two_avg = spender_two_df['Price'].mean()
spender_two_total = spender_two_df['Price'].sum()

spender_three_df = purchase_data[purchase_data['SN'] == top_spenders_df.iloc[2]['SN']]
spender_three_count = len(spender_three_df)
spender_three_avg = spender_three_df['Price'].mean()
spender_three_total = spender_three_df['Price'].sum()

spender_four_df = purchase_data[purchase_data['SN'] == top_spenders_df.iloc[3]['SN']]
spender_four_count = len(spender_four_df)
spender_four_avg = spender_four_df['Price'].mean()
spender_four_total = spender_four_df['Price'].sum()

spender_five_df = purchase_data[purchase_data['SN'] == top_spenders_df.iloc[4]['SN']]
spender_five_count = len(spender_five_df)
spender_five_avg = spender_five_df['Price'].mean()
spender_five_total = spender_five_df['Price'].sum()
```

```python
# Add the purchase count values to this series
top_spenders_count = [spender_one_count, spender_two_count, spender_three_count, spender_four_count, spender_five_count]
# Add series to column
top_spenders_df['Purchase Count'] = top_spenders_count

# Add the average purchase prices to this series
top_spenders_avg = [spender_one_avg, spender_two_avg, spender_three_avg, spender_four_avg, spender_five_avg]
# Add series to column
top_spenders_df['Average Purchase Price'] = top_spenders_avg

# Format the average purchase price column as currency
top_spenders_df['Average Purchase Price'] = pd.Series(["${:,.2f}".format(val) for val in top_spenders_df['Average Purchase Price']], index = top_spenders_df.index)
# Format the total purchase value column as currency
top_spenders_df['Total Purchase Value'] = pd.Series(["${:,.2f}".format(val) for val in top_spenders_df['Total Purchase Value']], index = top_spenders_df.index)

# Reorder the columns so that Total Purchase Value is the last column
top_spenders_df = top_spenders_df[['SN','Purchase Count','Average Purchase Price','Total Purchase Value']]
# Set the screen names as the index
top_spenders_df = top_spenders_df.set_index('SN')
# Print the final dataframe
top_spenders_df
```
